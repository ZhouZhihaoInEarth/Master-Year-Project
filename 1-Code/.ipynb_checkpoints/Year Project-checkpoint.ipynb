# 5. Defining Custom Dataset 

```python
class MedicalImageDataset(Dataset):  
    
    
    # Store the directory where the dataset to be read is located
    def __init__(self, imagePaths, maskPaths, transforms):
        
        # Store the image paths
        self.imagePaths = imagePaths
        
        # Store the image paths
        self.maskPaths = maskPaths
        
        # Transformations
        self.transforms = transforms
        
       
    #  Returns the total number of image paths in the dataset  
    def __len__(self):
        
        return len(self.imagePaths)
    
    
    
    # Returns the data at the corresponding location based on the index
    # Index starts from 0
    def __getitem__(self, index):
        
        # The image corresponding to the current index
        imagePath = self.imagePaths[index]
        
        # Return array(height，width，channel)
        # The order of channel is BGR
        image = cv2.imread(imagePath)
        # BGR --> RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Grayscale mode
        mask = cv2.imread(self.maskPaths[index], 0)
        
        
        if self.transforms is not None:
            # apply the transformations to both image and its mask
            image = self.transforms(image)
            mask = self.transforms(mask)  
        return (image, mask)
```
